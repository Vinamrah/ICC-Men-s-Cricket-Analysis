```python
# user header - identify the client making the request, ensuring compatibility and sometimes avoiding blocks by the server.
import requests
from bs4 import BeautifulSoup

url = "https://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;spanmax1=31+May+2024;spanmin1=01+Jan+1990;spanval1=span;template=results;type=bowling"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

response = requests.get(url, headers=headers)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    # Your scraping logic
else:
    print(f"Failed to retrieve content: {response.status_code}")
```

```python
content = soup.find_all("tr",class_='data1')
print(len(content))
```

```python
for i in content:
    # print(i) 
    player = i.find_all('td')
    # print(player)
    players_info = []
    for j in player:
        if(j.text!=''):
           # print(j.text)
            players_info.append(j.text)
    break
```

```python
# Finally scraping information from pages 1 to 30

# Initialize the final list to store players' information
final_players_list = []

# Loop through pages 1 to 30
for i in range(1, 31):
    url = f"https://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;page={i};spanmax1=31+May+2024;spanmin1=01+Jan+1990;spanval1=span;template=results;type=bowling"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Will raise an HTTPError for bad responses
    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve content from page {i}: {e}")
        continue

    soup = BeautifulSoup(response.content, 'html.parser')
    content = soup.find_all("tr", class_='data1')

    for k in content:
        player = k.find_all('td')
        players_info = [j.text.strip() for j in player if j.text.strip()]
        if players_info:
            final_players_list.append(players_info)

print(final_players_list)

```

```python
field = ['name','span','matches','innings','overs','maiden_overs','runs_conceded','wickets','best_bowling_inning','best_bowling_match','average','economy_rate','bowling_strike_rate','five_wickets_haul','ten_wickets_haul']
```

```python
import csv
with open("TEST_bowling_record.csv",'w') as f:
    csvwriter= csv.writer(f)
    csvwriter.writerow(field)
    csvwriter.writerows(final_players_list)
    
f.close()
```
